```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = [
    "JetBrainsMono Nerd Font Mono",
    "Product Sans",
    "Helvetica",
]
plt.rcParams["font.size"] = 14
```

```python
##load dataset iris.csv
iris_df = pd.read_csv("./datasets/iris.csv")
print("Dataset Overview:\n")
iris_df.head()
```

```python
# Display the shape of the dataset
print("Dimension of dataset -> ", iris_df.shape)
```

```python
# List all column names
print("Columns in the dataset:")
pd.DataFrame(iris_df.columns, columns=["Columns"])
```

```python
# Display summary statistics for numerical columns
print("\nStatistical description of dataset: ")
description = iris_df.describe(include=["float"])
np.round(description, 2)
```

```python
# Dataset information (data types, non-null counts, etc.)
print("\nDataset Information:")
iris_df.info()
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 10))

# Feature names and colors
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
colors = ["#FF828277", "#FFC88277", "#96FF9677", "#8282FF77"]

# Loop through each feature and create a histogram
for ax, feature, color in zip(axes.flat, features, colors):
    sns.histplot(
        data=iris_df,
        x=feature,
        bins=10,
        kde=True,
        color=color,
        edgecolor="black",
        ax=ax,
    )
    ax.set_title(f"Distribution of {feature.capitalize()}")  # Add title to each subplot

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 9))

# Feature names
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Loop through each feature and create a histogram
for ax, feature in zip(axes.flat, features):
    sns.histplot(
        data=iris_df,
        x=feature,
        hue="species",
        bins=10,
        kde=True,
        palette="plasma",
        ax=ax,
        edgecolor="black",
    )
    ax.set_title(f"{feature.replace('_', ' ').title()} Distribution")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.boxplot(
    data=iris_df.drop(["species"], axis=1),
    palette=["#FF828277", "#FFC88277", "#96FF9677", "#8282FF77"],
    linewidth=2,
)
plt.title("Boxplot of Iris Dataset")
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 9))

# Feature names and colors
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
colors = ["#FF828277", "#FFC88277", "#96FF9677", "#8282FF77"]

# Loop through each feature and create a boxplot
for ax, feature, color in zip(axes.flat, features, colors):
    sns.boxplot(
        data=iris_df,
        x="species",
        y=feature,
        color=color,
        linewidth=2,
        ax=ax,
    )
    ax.set_title(f"{feature.replace('_', ' ').title()} by Species")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 9))

# Feature names and colors
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
colors = ["#FF828277", "#FFC88277", "#96FF9677", "#8282FF77"]

# Loop through each feature and create a KDE plot
for ax, feature, color in zip(axes.flat, features, colors):
    sns.kdeplot(
        data=iris_df,
        x=feature,
        fill=True,
        color=color,
        alpha=0.5,
        ax=ax,
    )
    ax.set_title(f"{feature.replace('_', ' ').title()} Distribution")

plt.tight_layout()
plt.show()
```
